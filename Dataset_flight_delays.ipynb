```python
import pandas as pd
data= pd.read_csv("Dataset_flight_delays.csv")
print(data.shape)
data.head()
```

# Preprocessing


```python
#Null value treatment
data.isnull().sum()
```

# Replacing strings with numbers


```python
#Function for dictonary to store coding info about varaibles
info={}
def code_map(le,name):
    global info
    job_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    info[name]=job_name_mapping
    print(name,' :varaiable successfully coded')    
```

```python
data.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
#convert the categorical columns into numeric
data['Month']=le.fit_transform(data['Month'])
code_map(le,'Month')

data['DayofMonth']=le.fit_transform(data['DayofMonth'])
code_map(le,'DayofMonth')

data['DayOfWeek']=le.fit_transform(data['DayOfWeek'])
code_map(le,'DayOfWeek')

data['UniqueCarrier']=le.fit_transform(data['UniqueCarrier'])
code_map(le,'UniqueCarrier')

data['Origin']=le.fit_transform(data['Origin'])
code_map(le,'Origin')

data['Dest']=le.fit_transform(data['Dest'])
code_map(le,'Dest')

data.head()
```

```python
info
```

```python
X=data.iloc[:,1:9]
y=data.iloc[:,-1]
```

```python
#Spliting data
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2, random_state=1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators=300, random_state=1)
RFC.fit(X_train, y_train)
y_pred = RFC.predict(X_test)
print('Accuracy',accuracy_score(y_test,y_pred))
```

```python
#confusion matrix
from sklearn import metrics
conf_matr =metrics.confusion_matrix(y_test,y_pred)
conf_matr
```
